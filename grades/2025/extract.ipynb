```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Settings
input_file = Path("grades_2025.xlsx")
output_file = Path("grades_2025_extract.xlsx")
df = pd.read_excel(input_file, engine="openpyxl")
df.head()
```

```python
desired_columns = ["First name", "Last name", "ID number", "Email address", "Assignment 1", "Marker 1", "Assignment 2", "Marker 2", "Assignment 3", "Marker 3"]
ids_to_drop = [2577950, 2597473, 106206, 3035565, 334275, 2787095]
```

```python
df = df[desired_columns]
df = df.fillna(0)
```

```python
# Drop rows with specified ID numbers (coerce to numeric for comparison)
id_numeric = pd.to_numeric(df["ID number"], errors="coerce")
id_numeric
```

```python
df_to_drop = df.loc[id_numeric.isin(ids_to_drop)]
df_to_drop
```

```python
df_to_keep = df[~id_numeric.isin(ids_to_drop)]
df_to_keep.to_excel(output_file, index=False, engine="openpyxl")
print(f"Saved {len(df)} rows to {output_file}")
```
