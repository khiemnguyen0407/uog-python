```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
std_list = pd.read_excel("./grades_2025.xlsx")
```

```python
markers = ['Khiem', 'Fin', 'Niamh', 'Huy', 'Doris']
masks = np.array([[std_list[f'Marker {i + 1}'].to_numpy() == name 
                   for name in markers] for i in range(3)])
print(masks.dtype)
print(masks.shape)
```

```python
output_path = Path("statistics.txt")
```

```python
def print_statistics(filename=None):
    output_file = None
    if filename:
        output_path = Path(filename) if filename else (Path.cwd() / "statistics")
        
        output_file = output_path.open("e", encoding="utf-8")

    def write_output(text):
        if output_file:
            print(text, file=output_file)
        else:
            print(text)
    
    
    
    for iass, ass_num in enumerate(range(1, 4, 1)):
        write_output(f"\033[1mAssignment {ass_num}\033[0m")
        for iname, name in enumerate(markers):
            # Compute the average grade for one particular assignment

            ass_grades = std_list[f'Assignment {ass_num}'].to_numpy()
            current_mask = masks[iass, iname]
            # print(ass_grades.shape, current_mask.shape)
            total_grade = np.sum(ass_grades[current_mask])
            total_count = np.sum(current_mask)
            avg_grade = total_grade / total_count if total_count != 0 else 0
            write_output(f"{name:<5}: Average = {np.round(avg_grade, 4):<6} -- "
                         f"Total = {total_count}")
        write_output("")
            # print(f"Average Grade of Assignment {ass_num} marked by {name:<5} = {np.round(avg_grade, 4)}", file=f, end=" -- ")
            # print(f" Total marks = {total_count}", file=f)
    if output_file:
        output_file.close()
```

```python
print_statistics()
```
