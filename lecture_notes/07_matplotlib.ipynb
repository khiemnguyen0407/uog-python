# Intro to `matplotlib` and `maplotlib.pyplot`

***Matplotlib*** is a comprehensive library for creating static, animated, and iteractive visualization in Python. ***Matplotlib*** makes easy things easy and hard things possible.

- Create *publication quality plots*.
- Make interactive figures that can zoom, pan, update.
- Customize *visual style* and *layout*.
- Export to *many file fortmats*.
- Embed in *JupyterLab* and *Graphical User Interfaces*.
- Use a rich array of third-party packages built on Matplotlib

`matplotlib.pyplot` is a collection of functions that make matplotlib work like MATLAB. Each `pyplot` function makes some change to a figure: e.g., creates a figures, create a plotting area in a figures, plots some lines in ploting area, decorates the plot with labels, etc.

In `matplotlib.pyplot` various states are preserved across function calls, so that it keeps track of things like current figure and plotting area, and the plotting functions are directed to current axes (please note that "axes" here and in most places in this notebook refers to the *axes* **part of a figure** and not strict mathematical term for more than one axis).

We can plot a simple function quite very easily with

Let us plot three simple functions $f(x) = \frac{1}{2}x$, $g(x) = x^2$ and $h(x) = \exp(x)$.

### Plotting with data generated from `numpy` library

#### Why `numpy` library?

**Numpy** has powerful lists and arrays of doubles or floats. In fact, in practice `matplotlib.pyplot` is frequently used with `numpy` arrays instead of built-in Python lists. Some of arguments for using `numpy` library:
1. `numpy` arrays are designed for floating numbers, not like list for any data types. As for plots, we always need some sort of numeric values, not just any data types.
2. Operations on `numpy` arrays are built in C/C++ and executed extremely fast as compared to those computed on the built-in list. In fact, we need several `for` loops in dealing with lists while we don't need loops when dealing with `numpy` arrays most of the time.

***Further information for your own interest***

In fact, the `numpy` library is in C/C++ and designed to handle sophisticated floating operations (or numerical operations) on multi-dimensional arrays and multi-order tensors ([Wikipedia](https://en.wikipedia.org/wiki/Tensor)). It saves a lot of computational time because Python is a high-level programming language and it must do a lot of overheading for excuting for loop while C is a much lower-level programming and the C-code was compiled before being used.

**Note**  To understand the power of `numpy` library, we will revisit the library in the future lectures.

#### Using of `numpy.arange()` and `numpy.linspace()` for one-dimensional arrays

To define an array of evenly distanced grid points, we normally use the two following functions:
* `numpy.arange(<start>, <stop>, <step>)`
> generate an array spaning from `<start>` to `<stop>` with a predefined controlling step size `<step>`. ***Note***: The value `<stop>` is excluded from the array.

* `numpy.linspace(<start>, <stop>, <number-of-points>)`
> generate an array spanning from `<start>` to `<stop>` with a predefined number of grid points `<number-of-points>`. All grid points are evenly distanced from each other. ***Note***: The value `<stop>` is included in the array.

````python
# arange(<start>, <stop>, <step>)
x1 = np.arange(0.0, 10.0, 0.1)
# values from 0 to 10.0 in steps of 0.1, includes 0.0 and excludes 10.0
x1
# print(x1)  # Uncomment this statement to see the result in print()
````

````python
# linspace(<start>, <stop>, <number-of-points>)
x2 = np.linspace(0.0, 10.0, 15)
# values from 0 to 10.0 using 15 numbers equally distant from each other
# includes 0.0 and includes 10.0
x2
# print(x2)  # Uncomment this statement to see the result in print()
````

````python
# We can also change the width of line.
plt.plot(x, y, linestyle="-.", marker='o', linewidth=2.0)  
plt.show()
````

````python
# The color of line is specified by the 3 values [Red, Green, Blue] in the RGB color mode
plt.plot(x, y, color=[0.5, 0.6, 0.2], linewidth=2.0) 
plt.show()
# The color of line can be specified by a letter as well, e.g., "b" -> blue, "r" -> red, "k" -> black, "g" -> green, "y" -> yellow, "m" -> magenta and more.
````

### Comparison of the execution time for Python list and `numpy` array

As already mentioned above, the `numpy` array is designed for fast operations on the multi-dimensional arrays of numbers and thus always beat the operations on the built-in Python list of numbers. This example shows how we can achieve a speed up in computation up to 10 times by using `numpy` arrays instead of Python list. This comparison is performed on a very small set of data. Imagine how much we can achieve when working on the large data. The magnitude of improvement in speed can read hundreds or thousands time.

````python
import math as mt
import matplotlib.pyplot as plt
import numpy as np

import time

start_time = time.time()
x = []
y = []
n_points = 500
delta = 2*mt.pi / n_points
for i in range(n_points+1):
    x.append(i * delta)
    y.append(mt.sin(x[i]))
    
plt.plot(x, y)
print("Execution time using Python list: {0}".format(time.time() - start_time))

start_time = time.time()
x = np.linspace(0, 2*np.pi, n_points)
y = np.sin(x)
plt.plot(x, y)
print("Execution time using Numpy array: {0}".format(time.time() - start_time))
````

### Discontinuous plots

Some functions are inherently discontinous, and that's tricky. For instance, the bending moment of a simply supported beam with point load in the middle is a *V* shaped plot.

````python
def bending_moment(x, length, load):
    if x <= length/2:
        return -x * load / 2
    else:
        return -(length - x) * load / 2
````

````python
# Ploting this function will give an error
length, load = 10, 2
x = np.arange(0, length, 0.1)
y = bending_moment(x, length, load)     # This should give error.
````

````python
# There is a neat trick for this issue
length, load = 10, 2
x = np.arange(0, length, 0.1)
bm_vectorized = np.vectorize(bending_moment)
y = bm_vectorized(x, length, load)
plt.plot(x, y)
plt.show()
````

### List comprehensions

***List comprehensions*** provide a concise way to create lists. Common applications are to make new lists where each element is the result of some operations applied to each member of another sequence or iterable, or to create a subsequence of those elements that satisfy a certain condition.

A list comprehension consists of square brackets (i.e. []) containing an expression followed by a `for` clause, then zero or more `for` or `if` clause. The result will be a new listing resulting from evaluating the expression in the context of the `for` and `if` clauses which follow it.


So, we can *rewrite* the following snippet of code
```Python
new_list = []
for i in old_list:
    if filter(i):
        new_list.append(expression(i))
```
as the following list comprehension expression
```Python
new_list = [expression(i) for i in old_list if filter(i)]
```

````python
## Example 1
def expresion(i):
    return i * i
old_list = list(range(0, 10))
new_list = []

for i in old_list:
    if i > 3 and i < 8:
        new_list.append(expresion(i))
        
print(old_list)
print(new_list)

# Equivalently, we can use the following list comprehension
new_list = [expresion(i) for i in old_list if i > 3 and i < 8]
print(new_list)
````

A summary for **list comprehension**

1. It is a short and efficient way to make lists
2. It comprises of brackets containing an expression, a `for` clause, then zero or more `for`/`if` clauses
3. We can put in all kinds of objects in lists using comprehensions
4. The resulting list comes from evaluating the expression in the context of the `for`/`if` clauses.

````python
## Example 2

# We create a list of tuples using two for clauses and one if clause.
comp_list = [(x, y) for x in [1, 2, 3] for y in [3, 1, 4] if x != y]
# This list is equivalent to doing
combs = []
for x in [1, 2, 3]:
    for y in [3, 1, 4]:
        if x != y:
            combs.append((x, y))
            
print("Using list comprehension: \n", comp_list)
print("No list comprehension:\n", combs)
````

````python
## More examples
from IPython.lib.display import isdir
from numpy import number


x = [i for i in range(10)]
print("x =", x)

squares = [i**2 for i in range(10)]
print("squares =", squares)

lower_case = [x.lower() for x in ["A", "B", "C", "X", "Y", "Z", "ABC"]]
print("lower_case =", lower_case)

# pre-existing items satisfying certain condition
string = "Hello World 2 times and 3 times"
number_only = [x for x in string if x.isdigit()]
print("number_only =", number_only)

# Multiple arguments
array_sum = [x + y for x in [10, 30, 50] for y in [20, 40, 60]]
array_mult = [x * y for x in [1, 2, 3] for y in [4, 5, 6]]
print("array_sum =", array_sum)
print("array_mult =", array_mult)
````

````python
## More examples
# Matrix with only value of one
matrix_ones = [[1 for x in range(5)] for y in range (5)]
print("matrix_ones =", matrix_ones)

# Matrix with values determined by x * y
matrix_mult = [ [j * i for j in range(0, 5, 1)] for i in range(6, 10, 1)] # 4 rows, 5 columns
print("matrix_mult =", matrix_mult)

# We can print matrix out more in the matrix form
print("matrix_ones =")
[print(matrix_ones[i]) for i in range(5)]

print("matrix_mult =")
[print(matrix_mult[i]) for i in range(4)]
# Unfortunately, we have to accept the array of None as the output of this statement at this point.
````
